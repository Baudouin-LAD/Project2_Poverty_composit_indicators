# Project 2: composite poverty indicator

```python
import pandas as pd
import matplotlib.pyplot as plt

# read csv file and create dataframe
df = pd.read_csv('poverty_indicator.csv')
df = df.sort_values(by=['country_name', 'year'])

# set year as index
df.set_index('year', inplace=True)

# group by country
grouped = df.groupby('country_name')

# set colors for each country
colors = {'Denmark':'red', 'France':'blue', 'Germany':'black', 'Greece':'purple', 'Hungary':'orange', 'Portugal':'green', 'Sweden':'yellow', 'United Kingdom':'grey', 'United States':'navy'}

# create figure and axis
fig, ax = plt.subplots()

# loop through each country
for name, group in grouped:
    ax.plot(group['composite_poverty_indicator'], marker='o', linestyle='-', label=name, color=colors[name])

# add legend, title, and labels
plt.legend()
plt.title('Poverty Indicator by Country and Year')
plt.xlabel('Year')
plt.ylabel('Poverty Indicator')
ax.legend(bbox_to_anchor=(0, 0), loc='upper right', ncol=1)
# display graph
plt.show()
```
